# 2. On which Day food wastage percentage was Highest and Lowest?


### To find food waste percentage we will divide total attendence per day by Total meals served per day which is 700 for weekdays (Monday, Tuesday, Wednesday,Thursday, Friday)and 500 for weekends(Saturday and Sunday).

```python
df.loc[df['Days'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']), 'Total'] /= 700
df.loc[df['Days'].isin(['Saturday', 'Sunday']), 'Total'] /= 500
print(df)
```

```python
df = df.sort_values(by='Total', ascending=False)
df
```

```python
df["Total"] = df["Total"]*100
df
```

```python
df2 = df.groupby("Days")["Total"].mean().reset_index()
df2
```

### In Total column we have found out the amount of food consumed and to find food wastage percentage we have to subtract elements of Total column by 100 

```python
df2["Total"]= 100 - df2["Total"]
df2
```

### Thus now,Total column is showing food wastage percentage.

```python
import matplotlib.pyplot as plt

days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
Total = [22.51, 20.05, 3.53, 16.85, 13.5, 2.40, 4.05]

plt.bar(days, Total)
plt.xlabel('Days')
plt.ylabel('Food Wastage percentage')
plt.title('Food Waste percentage by Days')
plt.show()
```

```python
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
Total = [22.51, 20.05, 3.53, 16.85, 13.5, 2.40, 4.05]

plt.pie(Total, labels=days,autopct='%1.1f%%')

plt.title('Food waste Percentage By Days')

plt.show()
```

### Food wastage was maximum on Mondays (22.51%) and least on Saturdays (2.40%) 

# **Time to find suggestions to lessen food wastage** 

###  Mess is making 700 meals on weekdays and 500 meals on weekends constantly although attendence is way lesser.
### We can suggest how many meals should be made on different Days of the week respectively to drastically reduce food wastage

### We had found average attendance on different days of week which is 542.40 for monday, 559.60 for tuesday, 675.25 for wednesday, 582.00 for Thursday, 605.50 for friday, 488.00 for saturday and 479.75 for sunday.


### For Mondays, average attendence was 542.40 and the maximum attendence ever recorded on any monday was 550 so, we will add +5 meals of tolerance in case maximum increases. Thus , on Mondays 555 meals should be prepared constantly.

### Similarily, on Tuesdays 569, Wednesdays 688, Thursdays 591, Fridays 615, Saturdays 499 and on Sundays 488 meals should be prepared.

## Now, again finding food waste percentage after I suggested new amount of meals to be made on different days of week:

```python
df = pd.read_excel("Mess_attendence_report_June_month.xlsx")
df
```

```python
df["Total"]=df["breakfast"]+df["lunch"]+df["snacks"]+df["dinnner"]
df
```

### Now, I will divide total attendence on different Mondays by 555, Tuedays by 569, Wednesdays by 688, Thursdays by 591, Fridays by 615, Saturdays by 499, and Sundays by 488 to find new food consumption percentage.

```python
df.loc[df['Days'].isin(['Monday']), 'Total'] /= 555
df.loc[df['Days'].isin(['Tuesday']), 'Total'] /= 569
df.loc[df['Days'].isin(['Wednesday']), 'Total'] /= 688
df.loc[df['Days'].isin(['Thursday']), 'Total'] /= 591
df.loc[df['Days'].isin(['Friday']), 'Total'] /= 615
df.loc[df['Days'].isin(['Saturday']), 'Total'] /= 499
df.loc[df['Days'].isin(['Sunday']), 'Total'] /= 488

print(df)
```

### To find food wastage percentage we will subratract 100 by food consumption percentage.

```python
df["Total"]= 100 - df["Total"]
df
```

## Finding average of food waste percentages:

```python
df1 = df['Total'].mean()
df1
```

## Now, food waste percentage is down to only 1.82% only.
